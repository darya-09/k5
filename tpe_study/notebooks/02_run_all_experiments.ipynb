```python
# Зависимости (в Colab):
# !pip -q install numpy scipy pandas matplotlib optuna openpyxl
```

```python
# Настройка: находим корень проекта tpe_study и добавляем src в путь.
# В Colab сначала склонируйте репозиторий (раскомментируйте):
# !git clone -b claude/quirky-euler-e3laj1 https://github.com/darya-09/k5.git
# %cd k5/tpe_study
import sys, os, json
from pathlib import Path
ROOT = Path.cwd()
if not (ROOT/'src'/'tpe_study').exists():
    for c in [Path('tpe_study'), Path('k5/tpe_study'), Path('/content/k5/tpe_study')]:
        if (c/'src'/'tpe_study').exists(): ROOT=c; break
sys.path.insert(0, str((ROOT/'src').resolve()))
print('ROOT =', ROOT.resolve())
```

```python
# Полный прогон. Для быстрой проверки добавьте '--quick'.
import subprocess, sys
cmd = [sys.executable, str(ROOT/'run.py')]
env = dict(os.environ, PYTHONHASHSEED='0')
p = subprocess.run(cmd, cwd=str(ROOT), env=env, capture_output=True, text=True)
print(p.stdout[-3000:]);  print('STDERR:', p.stderr[-1500:] if p.returncode else 'ok')
```

```python
# Сгенерировать docs/RESULTS.md + xlsx из полученных таблиц.
p = subprocess.run([sys.executable, str(ROOT/'make_results_doc.py')], cwd=str(ROOT), env=env, capture_output=True, text=True)
print(p.stdout, p.stderr[-500:])
```
